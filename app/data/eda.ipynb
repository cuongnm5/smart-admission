```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 160)
sns.set_theme(style="whitegrid")
```

```python
csv_path = "universities_master.csv"
df = pd.read_csv(csv_path)

print(f"Shape: {df.shape}")
df.head()
```

```python
print("\nColumn types:")
print(df.dtypes.value_counts())

summary = pd.DataFrame({
    "dtype": df.dtypes.astype(str),
    "missing_count": df.isna().sum(),
    "missing_pct": (df.isna().mean() * 100).round(2),
    "n_unique": df.nunique(dropna=True)
}).sort_values(["missing_pct", "n_unique"], ascending=[False, False])

summary.head(25)
```

```python
duplicate_rows = df.duplicated().sum()
print(f"Duplicate rows: {duplicate_rows}")

missing_by_row = df.isna().sum(axis=1)
print("Rows with at least one missing value:", int((missing_by_row > 0).sum()))
print("Rows fully complete:", int((missing_by_row == 0).sum()))

missing_cols = df.isna().mean().sort_values(ascending=False)
missing_cols[missing_cols > 0].head(20)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numeric columns ({len(numeric_cols)}):", numeric_cols)

if numeric_cols:
    display(df[numeric_cols].describe().T)
else:
    print("No numeric columns detected.")
```

```python
plot_cols = numeric_cols[:6]
if plot_cols:
    n_cols = 3
    n_rows = int(np.ceil(len(plot_cols) / n_cols))
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 3.8 * n_rows))
    axes = np.array(axes).reshape(-1)

    for i, col in enumerate(plot_cols):
        sns.histplot(df[col].dropna(), kde=True, ax=axes[i], bins=30)
        axes[i].set_title(f"Distribution: {col}")

    for j in range(len(plot_cols), len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns available for plotting.")
```

```python
categorical_cols = [
    col for col in df.select_dtypes(include=["object", "category", "bool"]).columns
    if df[col].nunique(dropna=True) <= 30
]
print(f"Categorical columns (<=30 unique values): {len(categorical_cols)}")

for col in categorical_cols[:8]:
    print(f"\nTop values for {col}:")
    display(df[col].value_counts(dropna=False).head(10).to_frame("count"))
```

```python
if len(numeric_cols) >= 2:
    corr = df[numeric_cols].corr(numeric_only=True)
    plt.figure(figsize=(10, 7))
    sns.heatmap(corr, cmap="coolwarm", center=0)
    plt.title("Correlation Heatmap (Numeric Features)")
    plt.tight_layout()
    plt.show()
else:
    print("Not enough numeric columns to compute correlation.")
```

```python
candidate_metric_cols = [
    c for c in [
        "acceptance_rate",
        "tuition_in_state",
        "tuition_out_state",
        "room_and_board",
        "graduation_rate",
        "student_faculty_ratio"
    ] if c in df.columns
]

if candidate_metric_cols:
    metric_summary = df[candidate_metric_cols].describe().T
    display(metric_summary)
else:
    print("No expected university metrics found; review `df.columns` for available fields.")
```

```python
# Clean + export universities to JSONL with required fields
from pathlib import Path

required_output_cols = [
    "admissio_rate",
    "sat_avg",
    "tution_in_state",
    "tution_out_of_state",
    "programs_offered",
    "top_programs",
]

# Map expected output fields to possible input column variants
column_aliases = {
    "admissio_rate": ["admissio_rate", "admission_rate", "acceptance_rate"],
    "sat_avg": ["sat_avg", "sat_average", "avg_sat"],
    "tution_in_state": ["tution_in_state", "tuition_in_state", "in_state_tuition"],
    "tution_out_of_state": ["tution_out_of_state", "tuition_out_of_state", "out_of_state_tuition"],
    "programs_offered": ["programs_offered", "program_offered", "majors_offered"],
    "top_programs": ["top_programs", "top_program", "popular_programs"],
}

resolved_columns = {}
for out_col, aliases in column_aliases.items():
    matched = next((c for c in aliases if c in df.columns), None)
    if matched is None:
        raise KeyError(f"Missing expected column for {out_col}. Tried: {aliases}")
    resolved_columns[out_col] = matched

work_df = pd.DataFrame({
    out_col: df[src_col]
    for out_col, src_col in resolved_columns.items()
})

for col in required_output_cols:
    work_df[col] = work_df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)

non_empty_mask = pd.Series(True, index=work_df.index)
for col in required_output_cols:
    non_empty_mask &= work_df[col].notna()
    non_empty_mask &= work_df[col].astype(str).str.strip().ne("")
    non_empty_mask &= work_df[col].astype(str).str.lower().ne("nan")

clean_df = work_df.loc[non_empty_mask].copy()

output_path = Path("app/data/clean_universities.jsonl")
output_path.parent.mkdir(parents=True, exist_ok=True)
clean_df.to_json(output_path, orient="records", lines=True, force_ascii=False)

print(f"Source rows: {len(df)}")
print(f"Clean rows : {len(clean_df)}")
print(f"Dropped    : {len(df) - len(clean_df)}")
print(f"Saved file : {output_path}")
clean_df.head()
```

```python
import json

def read_jsonl(file_path):
    data = []
    try:
        with open(file_path, 'r') as f:
            for line in f:
                # Use json.loads() to parse each line (a string) into a Python dictionary
                json_object = json.loads(line.strip())
                data.append(json_object)
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
    return data
```
